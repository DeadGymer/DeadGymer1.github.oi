```python
import numpy as np
import matplotlib.pyplot as plt

xi = np.array([-3, -2, 1, 3])
fi = np.array([-4, 19, -8, 14])

def lagrange_interpolation(x, xi, fi):
    n = len(xi)
    result = 0.0
    for i in range(n):
        term = fi[i]
        for j in range(n):
            if j != i:
                term *= (x - xi[j]) / (xi[i] - xi[j])
        result += term
    return result

x_values = np.linspace(min(xi), max(xi), 1000)
y_values = lagrange_interpolation(x_values, xi, fi)

plt.figure(figsize=(8, 6))
plt.plot(xi, fi, 'ro', label='Дані точки')
plt.plot(x_values, y_values, label='Інтерполяційний багаточлен Лагранжа')
plt.title('Інтерполяційний багаточлен Лагранжа')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
